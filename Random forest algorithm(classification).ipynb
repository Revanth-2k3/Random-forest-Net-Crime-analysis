```python
# Preprocessing
# Handle missing values (if any)
data = data.dropna()
```

```python
# Define feature variables (X) and target variable (y)
# Replace 'target_column' with the column name you want to predict
target_column = "Country"  # Update this with the actual target column name
X = data.drop(columns=[target_column])
y = data[target_column]
```

```python

# Drop rows with missing or invalid values in critical columns
cleaned_data = data.dropna(subset=[
    "2023_Losses", "2022_Losses", "2021_Losses", "2024_Losses"
])
```

```python
# Create a target variable by classifying "2024_Losses" as 'High' or 'Low'
threshold = cleaned_data["2024_Losses"].median()  # Use median as the threshold
cleaned_data["Loss_Category"] = cleaned_data["2024_Losses"].apply(
    lambda x: "High" if x > threshold else "Low"
)
```

```python

# Define features and target
features = cleaned_data[[
    "2023_Losses", "2022_Losses", "2021_Losses", "2020_Losses", "2019_Losses"
]]
target = cleaned_data["Loss_Category"]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create and train the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_scaled, y_train)
```

```python
# Predict on the test set
y_pred = rf_classifier.predict(X_test_scaled)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
# Print results
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:\n")
print(report)
```
